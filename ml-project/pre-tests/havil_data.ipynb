```python
import pandas as pd
```

```python
df_meme = pd.read_csv('../data/processed/meme.csv')
df_ai = pd.read_csv('../data/processed/ai.csv')
df_rwa = pd.read_csv('../data/processed/rwa.csv')
df_gamming = pd.read_csv('../data/processed/gaming.csv')
```

```python
df_meme.head()
```

```python
df_meme['timestamp'] = pd.to_datetime(df_meme['timestamp'], unit='s')
df_ai['timestamp'] = pd.to_datetime(df_ai['timestamp'], unit='s')
df_rwa['timestamp'] = pd.to_datetime(df_rwa['timestamp'], unit='s')
df_gamming['timestamp'] = pd.to_datetime(df_gamming['timestamp'], unit='s')
```

```python
halving_dates = [pd.to_datetime('2012-11-28'), pd.to_datetime('2016-07-09'), pd.to_datetime('2020-05-11'), pd.to_datetime('2024-04-20'), pd.to_datetime('2028-03-28')]
```

```python
df_meme['days_until_halving'] = df_meme['timestamp'].apply(lambda x: min([(halving - x).days for halving in halving_dates if halving > x]) if any(halving > x for halving in halving_dates) else 0)
df_ai['days_until_halving'] = df_ai['timestamp'].apply(lambda x: min([(halving - x).days for halving in halving_dates if halving > x]) if any(halving > x for halving in halving_dates) else 0)
df_rwa['days_until_halving'] = df_rwa['timestamp'].apply(lambda x: min([(halving - x).days for halving in halving_dates if halving > x]) if any(halving > x for halving in halving_dates) else 0)
df_gamming['days_until_halving'] = df_gamming['timestamp'].apply(lambda x: min([(halving - x).days for halving in halving_dates if halving > x]) if any(halving > x for halving in halving_dates) else 0)
```

```python
df_meme.head()
```

```python
token_max = [
    df_meme.loc[df_meme['timestamp'].idxmax()],
    df_ai.loc[df_ai['timestamp'].idxmax()],
    df_rwa.loc[df_rwa['timestamp'].idxmax()],
    df_gamming.loc[df_gamming['timestamp'].idxmax()]
]
```

```python
for i, token in enumerate(['df_meme', 'df_ai', 'df_rwa', 'df_gamming']):
    print(f"Fila con el mínimo timestamp en {token}:\n")
    print(token_max[i])
    print("\n" + "-"*50 + "\n")
```
