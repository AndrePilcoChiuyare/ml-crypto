```python
import pandas as pd
```

```python
gaming = pd.read_csv('../data/processed/gaming.csv')
rwa = pd.read_csv('../data/processed/rwa.csv')
ai = pd.read_csv('../data/processed/ai.csv')
meme = pd.read_csv('../data/processed/meme.csv')

gaming.shape, rwa.shape, ai.shape, meme.shape
```

```python
# drop where marketcap is -1
gaming = gaming[gaming['marketcap'] != -1]
rwa = rwa[rwa['marketcap'] != -1]
ai = ai[ai['marketcap'] != -1]
meme = meme[meme['marketcap'] != -1]

gaming.shape, rwa.shape, ai.shape, meme.shape
```

```python
print(gaming['id'].nunique())
print(ai['id'].nunique())
print(meme['id'].nunique())
print(rwa['id'].nunique())
```

```python
gaming.describe().T
```

```python
gaming.drop_duplicates(subset='id', keep='first', inplace=True)
ai.drop_duplicates(subset='id', keep='first', inplace=True)
meme.drop_duplicates(subset='id', keep='first', inplace=True)
rwa.drop_duplicates(subset='id', keep='first', inplace=True)
gaming = gaming[['id', 'symbol', 'name', 'marketcap']]
ai = ai[['id', 'symbol', 'name', 'marketcap']]
meme = meme[['id', 'symbol', 'name', 'marketcap']]
rwa = rwa[['id', 'symbol', 'name', 'marketcap']]
```
