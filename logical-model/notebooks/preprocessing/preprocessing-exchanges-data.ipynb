# Adding extra exchanges extra information

```python
import pandas as pd
import numpy as np
import ast
```

```python
crypto = pd.read_csv('../../data/csv/crypto_information.csv')
print(crypto.shape)
crypto.head()
```

```python
exchanges = pd.read_csv('../../data/csv/exchanges_data_clean.csv')
print(exchanges.shape)
exchanges.head()
```

```python
exchanges.info()
```

```python
# Convert each string in the 'exchange_id' column to a list
crypto.loc[:, 'exchange_id'] = crypto['exchange_id'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
# Separate every item of the exchange list in a single row
exploded_df = crypto.explode('exchange_id')
exploded_df.head()
```

```python
# Merge with exchanges to get weekly_visits
merged_df = exploded_df.merge(exchanges, left_on='exchange_id', right_on='id', how='left')
merged_df.head()
```

```python
# Group by id_x and calculate sum and average of weekly_visits and spot_volume_usd
sums_df = merged_df.groupby('id_x').agg({
    'weekly_visits': ['sum', 'mean', 'max', 'min'],
    'spot_volume_usd': ['sum', 'mean', 'max', 'min']
}).reset_index()
```

```python
sums_df.columns = ['crypto_id', 'total_weekly_visits', 'average_weekly_visits', 'max_weekly_visits', 'min_weekly_visits', 'total_spot_volume_usd', 'average_spot_volume_usd', 'max_spot_volume_usd', 'min_spot_volume_usd']
sums_df.head()
```

```python
# Merge the sums back into the original crypto dataset
result_df = crypto.merge(sums_df, left_on='id', right_on='crypto_id', how='left')
result_df.drop(columns=['crypto_id'], inplace=True)
result_df.shape
result_df.head()
```

```python
result_df.info()
```

```python
result_df = result_df.fillna(0)
result_df.info()
```

```python
result_df.to_csv('../../data/csv/crypto_final.csv', index=False)
```
