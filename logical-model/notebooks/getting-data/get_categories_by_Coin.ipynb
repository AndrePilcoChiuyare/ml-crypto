# Categories of MarketCup

Categories API request

```python
#Librerias
import requests
import json
import csv
import os
```

```python
api_key = os.getenv('COINMARKETCAP_MARCO_KEY')

url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/categories'


# Encabezados de la petición
headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': api_key
}

# Hacer la petición GET con parámetros de consulta
response = requests.get(url, headers=headers)

# Verificar si la solicitud fue exitosa, sino sale un error
if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```

Save data of categories

```python
#Si fue exitosa la llamada de la API
if response.status_code == 200:
    data = response.json()
    print("Datos recibidos desde la API:")
    print(data)  # Imprime toda la respuesta
    
    # Asegurarte de que 'data' contiene una lista de categorías
    categories = data.get('data', [])
    if categories and isinstance(categories, list):
        #Guardamos en un csv las categorias
        filename = '../../data/csv/coin_categories.csv'
        # Abrir el archivo CSV para escribir
        with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
            # Definir los nombres de las columnas, esto en funcion a lo que nos dio la API y a lo que queremos guardar
            fieldnames = ['id', 'name', 'title', 'description', 'num_tokens', 'avg_price_change', 'market_cap', 'market_cap_change', 'volume', 'volume_change', 'last_updated']
            # Crear el escritor CSV
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            # Escribir los encabezados
            writer.writeheader()
            # Escribir las filas de datos
            for category in categories:
                # Verificar que cada categoría es un diccionario
                if isinstance(category, dict):
                    # Asegurarse de que la categoría tenga todas las claves necesarias
                    row = {field: category.get(field, '') for field in fieldnames}
                    writer.writerow(row)
                else:
                    print(f"Warning: Expected dict but got {type(category)}")
        print(f"Datos guardados en {filename}")
    else:
        print("Error: 'data' no contiene una lista de categorías o está vacío.")
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```

especific categories 

```python
#Obervamos que estas categorias son las principales para buscar, similares o las mismas categorias
filter_terms = [
    "Real World Assets",
    "RWA",
    "AI & Big Data",
    "memes",
    "gaming"
]
# Convertir los términos a minúsculas para una comparación sin distinción de mayúsculas, para la comparacion
filter_terms_lower = [term.lower() for term in filter_terms]
```

Filter especific categories

```python
if response.status_code == 200:
    data = response.json() 
    
    # Obtener la lista de categorías de la respuesta
    categories = data.get('data', [])
    
    # Filtrar las categorías que contienen alguno de los términos en 'name' o 'title' la palabras puesta en filter_terms
    filtered_categories = [
        category for category in categories
        if any(term in category.get('name', '').lower() or term in category.get('title', '').lower() for term in filter_terms_lower)
    ]
    
    # Mostrar las categorías filtradas
    if filtered_categories:
        for category in filtered_categories:
            print(f"Name: {category.get('name')}, id: {category.get('id')}")
    else:
        print("No categories matching the filter terms found.")
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```

save information of RWA coins

```python
import json
import csv
# Verificar si la solicitud fue exitosa
if response2.status_code == 200:
    data2 = response2.json()  # Convertir el contenido de la respuesta a JSON

    with open('../../data/json/coins_RWA.json', 'w') as json_file:
        json.dump(data2, json_file, indent=4)

    # Verificar si data['data'] es un diccionario que contiene la clave 'coins'
    if 'coins' in data2['data']:
        coins = data2['data']['coins']  # Acceder a la lista de monedas
    else:
        coins = []  # Si no existe 'coins', inicializamos una lista vacía

    # Definir el nombre del archivo CSV
    csv_filename = '../../data/csv/coins_RWA.csv'

    # Crear el archivo CSV y escribir los datos
    with open(csv_filename, mode='w', newline='', encoding='utf-8') as csv_file:
        fieldnames = ['id', 'name', 'symbol', 'slug', 'num_market_pairs', 'date_added', 'tags', 
                      'max_supply', 'circulating_supply', 'total_supply', 'cmc_rank', 
                      'price', 'volume_24h', 'percent_change_1h', 'percent_change_24h', 
                      'percent_change_7d', 'market_cap']
        
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()

        for coin in coins:
            quote = coin['quote']['USD']  # Obtener los datos financieros en USD
            writer.writerow({
                'id': coin['id'],
                'name': coin['name'],
                'symbol': coin['symbol'],
                'slug': coin['slug'],
                'num_market_pairs': coin['num_market_pairs'],
                'date_added': coin['date_added'],
                'tags': ','.join(coin['tags']),  # Convertir la lista de tags a una cadena
                'max_supply': coin.get('max_supply', 'N/A'),
                'circulating_supply': coin.get('circulating_supply', 'N/A'),
                'total_supply': coin.get('total_supply', 'N/A'),
                'cmc_rank': coin.get('cmc_rank', 'N/A'),
                'price': quote['price'],
                'volume_24h': quote['volume_24h'],
                'percent_change_1h': quote['percent_change_1h'],
                'percent_change_24h': quote['percent_change_24h'],
                'percent_change_7d': quote['percent_change_7d'],
                'market_cap': quote['market_cap']
            })
    print(f"Datos guardados exitosamente en 'data_criptos.json' y 'coins_criptos.csv'.")
else:
    print(f"Error: {response2.status_code}")
    print(response2.text)
```
