```python
import pandas as pd
import numpy as np
import json
import requests
import time
```

```python
# Function to load JSON
def loadJSON(filepath):
    with open(filepath) as file:
        return json.load(file)
```

```python
# Loading JSON
data = loadJSON('../../data/json/main-tokens-clean.json')
data
```

```python
# (There's no need for API key)

# Endpoint URL
url = f'https://api.coingecko.com/api/v3/coins/'

# Request header
headers = {
    'Accepts': 'application/json'
}
```

```python
# Counter to check the amount of tokens processed
counter = 0
for i, token in enumerate(data):
    # Adding the URI to the URL
    response = requests.get(f"{url}{token['gecko_id']}/market_chart?vs_currency=usd&days=365", headers=headers)

    # Checking if it's successful
    if response.status_code == 200:
        response_data = response.json()

        # Adding Market cap list to the JSON
        data[i]['Market cap'] = response_data  # Aqu se guarda la respuesta en 'Market cap'
        counter += 1
        
        print(f"{token['name']}: Successful")
    else:
        print(f"Error: {response.status_code}")
        print(response.text)

    time.sleep(2)

print('Successful requests:', counter)
```

```python
with open('../../data/json/tokens-tvl_prueba.json', 'w') as outfile:
    json.dump(data, outfile, indent=4)
    print("JSON successfully saved at '../../data/json/tokens-tvl_prueba.json'")
```
